```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/download.jpeg')
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # Convert from BGR (OpenCV format) to RGB

# 1. Convert the image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 2. Separate the color channels
blue_channel, green_channel, red_channel = cv2.split(image)

# 3. Create a figure with 3 subplots to visualize each color channel
fig, axes = plt.subplots(1, 3, figsize=(10, 5))

# Display the Red channel
red_img = np.zeros_like(image_rgb)  # Create a black image with the same size
red_img[:, :, 0] = red_channel  # Set the Red channel in RGB format
axes[0].imshow(red_img)
axes[0].set_title('Red Channel')
axes[0].axis('off')

# Display the Green channel
green_img = np.zeros_like(image_rgb)
green_img[:, :, 1] = green_channel  # Set the Green channel in RGB format
axes[1].imshow(green_img)
axes[1].set_title('Green Channel')
axes[1].axis('off')

# Display the Blue channel
blue_img = np.zeros_like(image_rgb)
blue_img[:, :, 2] = blue_channel  # Set the Blue channel in RGB format
axes[2].imshow(blue_img)
axes[2].set_title('Blue Channel')
axes[2].axis('off')

# Display the grayscale image
plt.figure(figsize=(5,5))
plt.imshow(gray_image, cmap='gray')
plt.title('Grayscale Image')
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Get the image dimensions (height and width)
rows, cols = image.shape[:2]

# Define the translation matrix
# Shifting the image by 100 pixels along the x-axis and 50 pixels along the y-axis
translation_matrix = np.float32([[1, 0, 100], [0, 1, 50]])

# Apply the translation using cv2.warpAffine
translated_image = cv2.warpAffine(image, translation_matrix, (cols, rows))

# Convert the translated image from BGR to RGB format for displaying with Matplotlib
translated_image_rgb = cv2.cvtColor(translated_image, cv2.COLOR_BGR2RGB)

# Display the translated image
plt.figure(figsize=(5, 5))
plt.imshow(translated_image_rgb)
plt.title('Translated Image (100px Right, 50px Down)')
plt.axis('off')  # Hide the axis labels
plt.show()
```

**TASK 3**

```python
import cv2
import matplotlib.pyplot as plt


# Obtain the image dimensions
height, width = image.shape[:2]

# Define the rotation center (center of the image)
center = (width // 2, height // 2)

# Initialize a list to store rotated images
rotated_images = []

# Define the angles to rotate
angles = [90, 180, 270, 360]

# Copy the image and rotate by specified angles
for angle in angles:
    # Get the rotation matrix
    rotation_matrix = cv2.getRotationMatrix2D(center, angle, 1.0)  # Scale factor is 1.0 (no scaling)

    # Perform the rotation
    rotated_image = cv2.warpAffine(image, rotation_matrix, (width, height))

    # Convert the rotated image from BGR to RGB format for displaying
    rotated_images.append(cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB))

# Display the rotated images
plt.figure(figsize=(15, 5))

for i, rotated_image in enumerate(rotated_images):
    plt.subplot(1, 4, i + 1)  # 1 row, 4 columns
    plt.imshow(rotated_image)
    plt.title(f'Rotated Image ({angles[i]})')
    plt.axis('off')  # Hide the axis labels

# Show the plot
plt.tight_layout()
plt.show()
```

**TASK 12**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt


# Convert the image to RGB for displaying
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Define kernels for different filtering operations
# Smoothing (Blurring) kernel (Average filter)
kernel_avg = np.ones((5, 5), np.float32) / 25

# Gaussian kernel
kernel_gaussian = cv2.getGaussianKernel(5, 1)  # 5x5 Gaussian kernel
kernel_gaussian = kernel_gaussian * kernel_gaussian.T  # Convert to 2D

# Sharpening kernel (Unsharp Mask)
kernel_sharpen = np.array([[0, -1, 0],
                            [-1, 5, -1],
                            [0, -1, 0]])

# Edge Detection kernel (Sobel operator)
kernel_sobel_x = np.array([[-1, 0, 1],
                            [-2, 0, 2],
                            [-1, 0, 1]])

kernel_sobel_y = np.array([[1, 2, 1],
                            [0, 0, 0],
                            [-1, -2, -1]])

# Apply linear filtering
# Smoothing
smoothed_image = cv2.filter2D(image, -1, kernel_avg)
gaussian_blurred_image = cv2.filter2D(image, -1, kernel_gaussian)

# Sharpening
sharpened_image = cv2.filter2D(image, -1, kernel_sharpen)

# Edge Detection
edges_x = cv2.filter2D(image, -1, kernel_sobel_x)
edges_y = cv2.filter2D(image, -1, kernel_sobel_y)

# Convert to float32 before calculating magnitude to avoid type issues
edges_x_float = np.float32(edges_x)
edges_y_float = np.float32(edges_y)

# Combine the Sobel results to get the overall edges
edges_combined = cv2.magnitude(edges_x_float, edges_y_float)

# Normalize the edges_combined for visualization
edges_combined = cv2.normalize(edges_combined, None, 0, 255, cv2.NORM_MINMAX)
edges_combined = np.uint8(edges_combined)

# Visualization
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(3, 3, 1)
plt.imshow(image_rgb)
plt.title('Original Image')
plt.axis('off')

# Smoothed Image (Average Filter)
plt.subplot(3, 3, 2)
plt.imshow(cv2.cvtColor(smoothed_image, cv2.COLOR_BGR2RGB))
plt.title('Smoothed Image (Average Filter)')
plt.axis('off')

# Gaussian Blurred Image
plt.subplot(3, 3, 3)
plt.imshow(cv2.cvtColor(gaussian_blurred_image, cv2.COLOR_BGR2RGB))
plt.title('Gaussian Blurred Image')
plt.axis('off')

# Sharpened Image
plt.subplot(3, 3, 4)
plt.imshow(cv2.cvtColor(sharpened_image, cv2.COLOR_BGR2RGB))
plt.title('Sharpened Image')
plt.axis('off')

# Edge Detection (Sobel X)
plt.subplot(3, 3, 5)
plt.imshow(cv2.cvtColor(edges_x, cv2.COLOR_BGR2RGB))
plt.title('Edges (Sobel X)')
plt.axis('off')

# Edge Detection (Sobel Y)
plt.subplot(3, 3, 6)
plt.imshow(cv2.cvtColor(edges_y, cv2.COLOR_BGR2RGB))
plt.title('Edges (Sobel Y)')
plt.axis('off')

# Combined Edges
plt.subplot(3, 3, 7)
plt.imshow(edges_combined, cmap='gray')
plt.title('Combined Edges')
plt.axis('off')

plt.tight_layout()
plt.show()
```
